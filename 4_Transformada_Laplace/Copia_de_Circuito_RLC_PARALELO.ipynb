# ***EJERCICIO***

 Encuentre la funci칩n de transferencia (demostraci칩n matem치tica) para un circuito RLC paralelo (entrada: fuente de corriente AC, salida: corriente en inductor), con condiciones iniciales cero. Luego, utilizando la API de control de sympy, simular y graficar:
1. Diagrama de polos y ceros.
2. Diagrama de bode.
3. Respuesta impulso.
4. Respuesta escal칩n.  
5. Respuesta rampa.

```python
import sympy as sym
import matplotlib.pyplot as plt
import sympy.physics.control as control

sym.init_printing()

s = sym.symbols('s', complex=True)
t, R, L, C, w = sym.symbols('t R L C omega', positive=True)
#Entrada de inter칠s en Laplace
X = w / (s**2 + w**2) # Ajustar seg칰n tipo de entrada
H=control.TransferFunction(1,(L*C)*s**2 + (L*s)/R + 1,s) #Funcion de transferencia con la API de control
#Hlc = H/(1+H) #para lazo cerrado

Y = H.num * X /H.den#funci칩n de transferencia del circuito multiplicada por X(s)
#Funci칩n de salida
Y
```

```python
#Modelo en el tiempo en funci칩n de RLC
Ye = Y
#Se recorre cada fracci칩n parcial seg칰n factorizaci칩n para simplificar c치lculos
for j in range(len(Ye.args)):
    if j==0:#primer elemento de la solucion
      yt = sym.inverse_laplace_transform(Ye.args[j], s, t)
    else: #siguientes elementos concatenados
      yt+= sym.inverse_laplace_transform(Ye.args[j], s, t)
yt
```

```python
#Modelo en el tiempo con RLC evaluados
RLC = {R: 1, L: sym.Rational('.5'), C: sym.Rational('.4')}#valores utilizados para la simulaci칩n
Ye = Y.subs(RLC)
for j in range(len(Ye.args)):
    if j==0:
      yt = sym.inverse_laplace_transform(Ye.args[j], s, t)
    else:
      yt+= sym.inverse_laplace_transform(Ye.args[j], s, t)

yt
```

```python
#Funci칩n de salida Y(s) -- Laplace
Ye
```

```python
#Funci칩n de salida y(t) -- en tiempo
yt
```

#1. Diagrama de polos y ceros.

```python
poles = sym.roots(sym.denom(Ye), s) #se obtienen la raices del denominador
poles
```

```python
zeros = sym.roots(sym.numer(Ye), s) #ra칤ces del numerador
zeros
```

```python
HRLC=control.TransferFunction(H.num,H.den.subs(RLC),s)
control.pole_zero_plot(HRLC)
```
